## Baseline DQN

````python
import gymnasium as gym
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.evaluation import evaluate_policy

# Create environment
env = make_atari_env("ALE/DemonAttack-v5", n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)

# Hyperparameters 
total_timesteps = 100_000
test_episodes = 20
````

### DQN Model Training Configuration

A Deep Q-Network (DQN) is initialized with a convolutional neural network (CnnPolicy) to handle image-based input from the Atari environment. Key hyperparameters include:

- **Learning Rate**: `0.0001` — Controls how quickly the model updates.
- **Gamma**: `0.99` — High discount factor for long-term rewards.
- **TensorBoard Logging**: Enabled for monitoring training progress.

The model is trained for `total_timesteps` and saved as `"dqn_demonattack"` for later evaluation or inference.


````python
model = DQN(
    "CnnPolicy",
    env,
    learning_rate=0.0001,
    gamma=0.99,
    verbose=1,
    tensorboard_log="./dqn_demonattack_tensorboard/"
)

model.learn(
    total_timesteps=total_timesteps,
    tb_log_name="dqn_cleanrun"
)
model.save("dqn_demonattack")
````

## Inspecting Q-Values from Trained DQN Agent

The following code snippet demonstrates how to extract Q-values from a trained DQN model for a given environment observation in `DemonAttack-v5`.


````python
import torch

# Reset the environment to get the initial observation
obs = env.reset()

# Load your trained model (optional)
from stable_baselines3 import DQN
model = DQN.load("dqn_demonattack")

# Convert obs to tensor and permute to match CNN input
obs_tensor = torch.tensor(obs).float().permute(0, 3, 1, 2).to(model.device)

# Get Q-values
with torch.no_grad():
    q_values = model.q_net(obs_tensor)
    q_values_np = q_values.cpu().numpy()

# Print Q-values and best action
print("📈 Q-values for each action:", q_values_np)
print("🤖 Chosen action:", np.argmax(q_values_np))
````

Action 0 has the highest expected reward, so the agent selects it. This shows how the DQN uses its neural network to pick the best action based on learned Q-values.

## 🎥 Recording Gameplay of Trained DQN Agent

The following script captures video of the trained DQN agent playing **DemonAttack-v5** using the `RecordVideo` wrapper from Gymnasium.

````python
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack
from stable_baselines3.common.env_util import make_atari_env
from gymnasium.wrappers import RecordVideo
import os

# Load trained model
model = DQN.load("dqn_demonattack")

# Set up video folder
video_folder = "./videos"
os.makedirs(video_folder, exist_ok=True)

# Create raw Atari env
base_env = make_atari_env("ALE/DemonAttack-v5", n_envs=1, seed=0)
env = base_env.envs[0]  # Get raw gym.Env

# Wrap with RecordVideo before vectorizing
env = RecordVideo(env, video_folder=video_folder, name_prefix="dqn_demo", episode_trigger=lambda ep: True)

# Re-wrap in vectorized and stacked format
vec_env = DummyVecEnv([lambda: env])
vec_env = VecFrameStack(vec_env, n_stack=4)

# Record 3 episodes
for ep in range(3):
    obs = vec_env.reset()
    done = False
    total_reward = 0
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, _ = vec_env.step(action)
        total_reward += reward[0]
    print(f"🎮 Episode {ep+1} Reward: {total_reward}")

vec_env.close()
````

## Base DQN with Adjustable Hyperparameters (ε-greedy)
This configuration applies an adjustable ε-greedy exploration policy to train a DQN agent on the **DemonAttack-v5** Atari environment.
### 🔧 Hyperparameter Breakdown

| Parameter                 | Value        | Explanation |
|--------------------------|--------------|-------------|
| `learning_rate`          | `0.0005`     | Controls how quickly the model updates Q-values. A slightly higher value than baseline (`0.0001`) speeds up learning but may risk instability. |
| `gamma` (discount factor)| `0.90`       | Determines the importance of future rewards. A lower gamma (compared to `0.99` in baseline) favors immediate rewards and accelerates convergence. |
| `exploration_initial_eps`| `1.0`        | Sets the starting epsilon value. The agent begins by exploring randomly. |
| `exploration_final_eps`  | `0.05`       | Minimum epsilon value. As training progresses, the agent shifts from exploration to exploitation. |
| `exploration_fraction`   | `0.2`        | Indicates that epsilon will decay from 1.0 to 0.05 over the first 20% of training timesteps (20,000 out of 100,000). After that, the agent mostly exploits learned policies. |
| `total_timesteps`        | `100,000`    | Total number of frames the agent interacts with the environment. |
| `tb_log_name`            | `"dqn_epsilon"` | TensorBoard log name to track performance metrics for this configuration. |


````python
from stable_baselines3 import DQN
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy

# Create and wrap the environment
env = make_atari_env("ALE/DemonAttack-v5", n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)

# ε-greedy DQN with adjustable params
model = DQN(
    "CnnPolicy",
    env,
    learning_rate=0.0005,   
    gamma=0.90,            
    exploration_fraction=0.2,     # Portion of training with epsilon decay
    exploration_initial_eps=1.0,  # Starting ε
    exploration_final_eps=0.05,   # Minimum ε
    verbose=1,
    tensorboard_log="./dqn_demonattack_tensorboard/"
)

total_timesteps=100_000
model.learn(
    total_timesteps=total_timesteps,
    tb_log_name="dqn_epsilon"
)
model.save("dqn_epsilon")
````

## QRDQN Agent with NoisyNet for DemonAttack-v5

This configuration uses **QRDQN** (Quantile Regression DQN) with **NoisyNet** to train on the **DemonAttack-v5** Atari environment. Instead of ε-greedy exploration, NoisyNet injects learnable noise into network weights to drive exploration.

### 🔧 Hyperparameter Breakdown

| Parameter                   | Value        | Explanation |
|----------------------------|--------------|-------------|
| `learning_rate`            | `0.0005`     | Enables faster Q-value updates. Higher than the baseline DQN rate (`0.0001`) to encourage more responsive learning. |
| `gamma` (discount factor)  | `0.99`       | Emphasizes long-term rewards, similar to the baseline configuration. |
| `exploration_initial_eps` | `0.0`        | No random exploration via epsilon. NoisyNet handles all exploration implicitly. |
| `exploration_final_eps`   | `0.0`        | Remains zero since ε-greedy is disabled. |
| `exploration_strategy`    | `NoisyNet`   | Exploration is achieved by adding noise to the neural network’s weights during training. |
| `total_timesteps`         | `100,000`    | Total environment interaction steps used for training. |
| `tb_log_name`             | `"qrdqn_noisynet"` | TensorBoard run name for monitoring training progress and reward trends. |

---

**Why NoisyNet?**  
Unlike ε-greedy policies that decay over time, NoisyNet facilitates **state-dependent exploration**. The injected noise in linear layers helps the agent explore intelligently without needing manually tuned decay schedules.

**Result Summary:**  
This agent showed competitive performance and faster reward growth during early episodes compared to ε-greedy DQN, though final reward stability may vary depending on noise variance.



````python
from sb3_contrib import QRDQN
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack

# Env setup
env = make_atari_env("ALE/DemonAttack-v5", n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)

# QRDQN with NoisyNet (ε-greedy disabled)
model_noisy = QRDQN(
    "CnnPolicy",
    env,
    learning_rate=0.0005,
    gamma=0.99,
    exploration_initial_eps=0.0,
    exploration_final_eps=0.0,
    verbose=1,
    tensorboard_log="./dqn_demonattack_tensorboard/"
)

# Train for full steps
model_noisy.learn(
    total_timesteps=100_000,
    tb_log_name="qrdqn_noisynet"
)
model_noisy.save("qrdqn_noisynet")
````
